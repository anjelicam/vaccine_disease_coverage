# Functions for tidying data

```python
def tidy_coverage(vaccine):
    # pivots year values to its own column called year, creates coverage percentage column
    melt = pd.melt(vaccine,
                   id_vars = ['country', 'vaccine'],
                   var_name = 'year',
                   value_name = 'coverage_percentage')
    
    # sorts data by country and year
    sort = melt.sort_values(by = ['country', 'year'])
    
    return sort

def tidy_cases(vaccine):
    # pivots year values to its own column called year, creates cases column
    melt = pd.melt(vaccine,
                   id_vars = ['country', 'vaccine'],
                   var_name = 'year',
                   value_name = 'cases')
    
    # sorts data by country and year
    sort = melt.sort_values(by = ['country', 'year'])
    
    return sort
```

```python
def tidy_data(df, value_column_name):
    """pivots year values to its own column identified by value_column_name"""
    return pd.melt(df,
                   id_vars = ['country', 'vaccine'],
                   var_name = 'year',
                   value_name = value_column_name) \
             .sort_values(by = ['country', 'year'])
```
